```python
#alternative method to find the ori
#concentation of c in forward strand
import matplotlib.pyplot as plt
import numpy as np
gene = "CATGGGCATCGGCCATACGCC"
def sequence_to_list(seq):
    lst = np.empty(0)
    for i in seq:
        lst = np.append(lst, i, axis = None)
    print(lst.shape)
    return lst

def skew_plot(seq):
    value = 0
    values = np.empty(0)
    for i in seq:
        if i == 'G' or i == 'g':
            value += 1
            values = np.append(values, value, axis = None)
        elif i == 'C' or i == 'c':
            value -= 1
            values = np.append(values, value, axis = None)
        elif i == 'A' or i == 'a' or i == 'T' or i == 't':
            value += 0
            values = np.append(values, value, axis = None)
        else:
            print("invalid entry at " + i)
            break
    print(values.shape)
    return values
g = skew_plot(gene)
print(g)

lst_seq = sequence_to_list(gene)
print(lst_seq)
plt.plot(g)
```

```python
gene_1 = "TAAAGACTGCCGAGAGGCCAACACGAGTGCTAGAACGAGGGGCGTAAACGCGGGTCCGAT"
g_1 = skew_plot(gene_1)
print(g_1)

plt.plot(g_1)
```
